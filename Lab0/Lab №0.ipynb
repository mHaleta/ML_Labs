# Препроцесинг

```python
for i in range(df["Country"].values.shape[0]):
    df["Country"].values[i] = df["Country"].values[i].strip()
```

```python
df["Winning Percentage"] = np.array(df["Winning Percentage"].values, dtype="<U5").astype(float)
```

```python
values = np.array(df["Career Earnings"].values, dtype=str)
for i in range(values.shape[0]):
    if values[i] != "nan":
        values[i] = values[i][1:]
df["Career Earnings"] = np.array(values, dtype=float)
```

```python
values = np.array(df["Singles Record (Career)"].values, dtype=str)
for i in range(values.shape[0]):
    if values[i] != "nan":
        string = ""
        for ch in values[i]:
            if ch == "-":
                break
            else:
                string += ch
        values[i] = string
df.insert(4, "Number of Wins", np.array(values, dtype=float))
```

```python
avg_ranks = df.groupby("Country", as_index=False)["Country", "Rank"].mean()
avg_wins = df.groupby("Country", as_index=False)["Country", "Number of Wins"].mean()

ranks = avg_ranks["Rank"].values
wins = avg_wins["Number of Wins"].values
countries = avg_ranks["Country"].values
```

```python
plt.figure(figsize=(17, 8))
ax = plt.subplot()

ax.bar(np.arange(countries.shape[0])-0.15, ranks, width=0.25, label="Average ranks")
ax.bar(np.arange(countries.shape[0])+0.15, wins, width=0.25, label="Average wins")

plt.xticks(range(countries.shape[0]), countries, rotation=90)
plt.legend()
plt.grid()
plt.show()
```
